```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('/content/3.12.+Example.csv')
data.head()
```

```python
plt.scatter(data.Satisfaction, data.Loyalty)
plt.show()
```

```python
x = data.copy()
```

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
clusters = x.copy()
clusters['Clusters'] = kmeans.fit_predict(x)
clusters.head()
```

```python
plt.scatter(clusters['Satisfaction'], clusters['Loyalty'], c=clusters['Clusters'], cmap='rainbow')
plt.show()
```

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
x_scaled
```

```python
wcss = []
for i in range(1,11):
  kmeans = KMeans(i)
  kmeans.fit(x_scaled)
  wcss_iter = kmeans.inertia_
  wcss.append(wcss_iter)
```

```python
plt.plot(range(1,11), wcss)
plt.show()
```

```python
kmeans_new = KMeans(5)
kmeans_new.fit(x_scaled)
clusters_new = x.copy()
clusters_new['Clusters'] = kmeans_new.fit_predict(x_scaled)
clusters_new.head()
```

```python
plt.scatter(clusters_new['Satisfaction'], clusters_new['Loyalty'], c=clusters_new['Clusters'], cmap='rainbow')
plt.show()
```
